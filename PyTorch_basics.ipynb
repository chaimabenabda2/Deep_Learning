## Calcul du gradient avec autograd

```python
# Tenseur rempli de valeurs aléatoires issus d'une distribution normale centrée réduite
x = torch.randn(3, requires_grad = True)
print(x)
```

```python
# Autograd permet le calcul automatique de gradients d’une fonction par rapport à ses paramètres
# Elle sert ainsi à effectuer facilement l’étape de backpropagation lors de l’entraînement d’un réseau de neurones
# En effet, lorsqu’un tenseur est créé avec l’option require_grad = True, ceci signale à Autograd que chaque opération effectuée sur ce tenseur 
# doit être gardée en mémoire.
```

```python
y = x + 2
print(y)
```

```python
z = y*y*2
print(z)
```

```python
t = y.mean()
print(t)
```

```python
# différentiation automatique
```

https://pytorch.org/tutorials/beginner/basics/autogradqs_tutorial.html

```python
z.backward(torch.tensor([0.1,1.0,0.001])) # dz/dx
print(x.grad)
```

```python
weights = torch.ones(4, requires_grad = True)
for epoch in range(1):
    model_output = (weights * 3).sum()
    model_output.backward()
    print(weights.grad)
```

```python
weights = torch.ones(4, requires_grad = True)
for epoch in range(2):
    model_output = (weights * 3).sum()
    model_output.backward()
    print(weights.grad)
```

```python
weights = torch.ones(4, requires_grad = True)
for epoch in range(2):
    model_output = (weights * 3).sum()
    model_output.backward()
    print(weights.grad)
    weights.grad.zero_()
```

1. Forward pass: compute Loss
2. Compute local gradients
3. Backward pass: compute $\dfrac{\partial Loss}{\partial Weights}$ using the chain rule

```python
Image('figures/steps.png')
```

```python
Image('figures/example.png')
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.0)
w = torch.tensor(1.0, requires_grad = True)

# Forward pass and compute loss
y_hat = w*x
loss = (y_hat - y)**2
print(loss)
```

```python
# backward pass
loss.backward()
print(w.grad)
```
